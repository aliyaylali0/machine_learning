```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("audi.csv")
```

```python
df.head(3)
```

```python
df = df.drop(columns=["index","href","PPY","MileageRank","PriceRank","PPYRank","Score"])
```

```python
df = df.drop(columns=["Type"])
```

```python
df.head(3)
```

```python
df.columns=["yil","mil","motor","guc","vites","yakit","sahip","fiyat"]
```

```python
df.head(3)
```

```python
df["motor"] = df["motor"].str.replace("L"," ")
```

```python
df["motor"] = pd.to_numeric(df["motor"])
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.head(3)
```

```python
df = pd.get_dummies(df,columns=["vites","yakit"],drop_first=True)
```

```python
y = df[["fiyat"]]
x = df.drop("fiyat",axis=1)
```

```python
lm = LinearRegression()
model = lm.fit(x,y)
```

```python
df.head(3)
```

```python
df["yil"].min()
```

```python
model.predict([[2014,100000,1.6,120,3,0,0]])
```
